```
def distanceLineFromOrigin (a,b,c):
  import math
  dist = abs(c)/math.sqrt(a*a+b*b)
  return dist
```

```
distanceLineFromOrigin(1,0,-50) #x=50
```

```
distanceLineFromOrigin(0,1,-25)
```

```
def distanceparallelLine (a,b,c1,c2):
  import math
  dist = abs(c1-c2)/math.sqrt(a*a+b*b)
  return dist
```

```
distanceparallelLine (1,0,-10,-5)
```

```
#tanh
x=np.linspace(start =-10,stop =10,num=200)
y= (np.exp(x)-np.exp(-x))/(np.exp(-x)+np.exp(x))
%matplotlib inline
plt.plot(x,y)
```

```
#gaussian function
x=np.linspace(start =-10,stop =10,num=200)
y= np.exp(-(x**2))
%matplotlib inline
plt.plot(x,y)
```

```
#Rectified Linear Unit(ReLU) function
x=np.linspace(start =-10,stop =10,num=200)
y= np.maximum(0,x)
%matplotlib inline
plt.plot(x,y)
```

```
#Leaky ReLU function
x=np.linspace(start =-10,stop =10,num=200)
y= np.maximum(0.1*x,x)
%matplotlib inline
plt.plot(x,y)
```

```
#step function
x=np.linspace(start =-10,stop =10,num=200)
y=np.where(x>=0,1,0)
%matplotlib inline
plt.plot(x,y)
```

```
#cicle
#r=10 and center origin
r=10
x = np.linspace(start=-r, stop=r, num=r*10 )
y_plus= np.sqrt(r*r-x*x)
y_minus= -np.sqrt(r*r-x*x)
plt.plot(x,y_plus)
plt.plot(x,y_minus)
```

```
#square
#x=0 y=0 x=1 y=1
x1= np.linspace(start=0,stop=1,num=100)
y1= 0*x1
y2 = np.linspace(start=0,stop=1,num=100)
x2 = 0*y2
x3= np.linspace(start=0,stop=1,num=100)
y3= 0*x1+1
y4 = np.linspace(start=0,stop=1,num=100)
x4 = 0*y2+1
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
```

```
#rectangle
#x=0 y=0 x=2 y=1
x1= np.linspace(start=0,stop=2,num=100)
y1= 0*x1
y2 = np.linspace(start=0,stop=1,num=100)
x2 = 0*y2
x3= np.linspace(start=0,stop=2,num=100)
y3= 0*x1+1
y4 = np.linspace(start=0,stop=1,num=100)
x4 = 0*y2+2
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
```

```
#ellipse
#a=2 b=1 center origin
a=2
b=1
x= np.linspace(start=-a,stop=a,num=a*100)
y_plus= np.sqrt(1-((x*x)/(a*a)))*b
y_minus= -np.sqrt(1-((x*x)/(a*a)))*b
plt.plot(x,y_plus)
plt.plot(x,y_minus)
```

```
# plot a plane with a=1 b=1 c=0 d=10
# Plotting a 3D surface
from mpl_toolkits.mplot3d import Axes3D # Importing this is necessary when you plot in 3D
a = 1
b = 1
c = 0
d = 10
##############################
xx = np.linspace(-10, 10, 20)
zz = np.linspace(-10, 10, 20)
X, Z = np.meshgrid(xx, zz)
yy = (-a*X - c*Z - d) * 1.0 / b
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, yy, Z)
```

```
#Question: Can you plot the surface of a 2D Gaussian function on a 3D plot?
from mpl_toolkits.mplot3d import Axes3D
xx = np.linspace(-10, 10, 100)
yy = np.linspace(-10, 10, 100)
x, y = np.meshgrid(xx, yy)
zz = np.exp(-(x**2+y**2))
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(x, y, zz)
```
