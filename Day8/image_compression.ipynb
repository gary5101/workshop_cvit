```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import plotly.graph_objects as go
```

```
flag=cv2.imread('flag.jpg',cv2.IMREAD_GRAYSCALE)
cv2_imshow(flag)
```

```
number_of_int=flag.shape[0]*flag.shape[1]
number_of_int
```

```
u, s, vT = np.linalg.svd(flag,full_matrices=False)
d=np.diag(s)
print(s.shape)
reconstruct=u@d@vT
cv2_imshow(reconstruct)
```

```
u, s, vT = np.linalg.svd(flag,full_matrices=False)
fig=go.Figure(go.Scatter(x = np.array(range(1,len(s))),y = s))
fig.show()
components_used=10
s[components_used:]=0
reconstruct=u@np.diag(s)@vT
cv2_imshow(np.hstack([flag,reconstruct]))
used_integers=u.shape[0]*components_used+components_used+components_used*vT.shape[1]
compression=number_of_int/used_integers
compression
```
