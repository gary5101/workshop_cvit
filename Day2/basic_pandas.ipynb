```
import pandas as pd
import numpy as np
```

```
base_url='https://gist.githubusercontent.com/alti-tude/710cb9d4dfc7ebcd0afb9cf93d8f6a8d/raw/574730ba009e69b81d6c79f2fef2c3dd5145db0a/'
```

```
#load data
music=pd.read_csv(base_url+'MusicTop50.csv', index_col=0)
```

```
#visualize
music.head()
```

```
music.tail()
```

```
# clean data
#make column names proper
music.columns=['track_name','artist_name','genre','beats_per_minute','energy','danceability(db)','loudness','liveness','valence()','length','acousticness','speechiness','popularity']
music.info()
```

```
#handle null values
#by dropping rows
music.dropna().info()
```

```
#by replacing it with mean
music_mean=music.mean()
music.fillna(music_mean).head()
```

```
#by replacing it with median
music_median=music.median()
music.fillna(music_median).head()
```

```
music[music['artist_name']==music.sort_values(by=['popularity']).iloc[0]['artist_name']].sort_values(by=['loudness'],ascending=False).iloc[0]['track_name']
```
